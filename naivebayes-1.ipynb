```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
# 1. 加载原始数据对象
wine = load_wine()

# 2. 将数值矩阵 (X) 转为 DataFrame，并贴上对应的特征标签
df = pd.DataFrame(wine.data, columns=wine.feature_names)

# 3. 将分类结果 (y) 作为一个新列加入表格
df['target'] = wine.target
```

```python
df.head()
```

Use wine dataset from sklearn.datasets to classify wines into 3 categories. Load the dataset and split it into test and train. After that train the model using Gaussian and Multinominal classifier and post which model performs better. Use the trained model to perform some predictions on test data.

```python
df.groupby('target').size()
```

```python
df1 = df[['alcohol', 'malic_acid', 'color_intensity', 'od280/od315_of_diluted_wines','target']]
df1.head()
```

```python
inputs = df1.drop('target', axis=1)
target = df1['target']
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
X_test[:10]
```

```python
y_test[:10]
```

```python
model.predict(X_test[:10])
```

```python
model.predict_proba(X_test[:10])
```

```python
from sklearn.naive_bayes import MultinomialNB
model2 = MultinomialNB()
model2.fit(X_train, y_train)
```

```python
model2.score(X_test, y_test)
```

```python
model2.predict(X_test[:10])
```

```python
model2.predict_proba(X_test[:10])
```
